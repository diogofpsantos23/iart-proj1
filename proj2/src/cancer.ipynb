# Boxplot - Interquartile range (IQR)

A boxplot with 'diagnosis' on the x-axis and 'radius_mean' on the y-axis. Each box in the boxplot represents the interquartile range (IQR) of 'radius_mean' for each 'diagnosis', with the line in the box representing the median. The whiskers represent values within 1.5 times the IQR, and any points outside of this range are considered outliers and are plotted as individual points.

```python
# Create boxplot
sb.boxplot(x='diagnosis', y='radius_mean', data=df)

# Show the plot
plt.show()
```
